## Exploratory Data Analysis

In this section we will go over the aforementioned variables from our research question and explore wheter or not there is significant evidence that either the brand country, flavor vareity, or packaging style contributes toward the corresponding instant ramen brand's rating.

## 1. Brand Country

First, we will explore if theres a signficant difference in the number of ramen brands originating from different countries. 

Here, we're creating a Null hypothesis to determine if theres a relationship between countries and the ramen brands originating from them. 

#### Hypothesis
Null: There is a significance difference in the amount of ramen brands originating from different Countries.

Alternative: There is not a significance difference in the amount of ramen brands originating from different Countries.

```python
contingency_table = pd.crosstab(index=final_ramen['Brand'], columns=final_ramen['Country'])

# Perform chi-square test for independence
chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table)

# Interpret Results
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is significant evidence that there is a difference in the number of ramen brands originating from different countries.")
else:
    print("Fail to reject null hypothesis: There is no significant evidence to support a difference in the number of ramen brands originating from different countries.")
```


We can then infer that there is a significant difference between countries and the number of ramen brands produced within them.

Let's make a visualization for the distribution of number of ramen brands per country:

```python
#Combine data with USA and United States together
final_ramen['Country'] = final_ramen['Country'].replace('USA', 'United States')
ramen_counts = final_ramen.groupby('Country')['Brand'].nunique().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
colors = sns.color_palette("rocket", len(ramen_counts))
ramen_counts.plot(kind='bar', color=colors)


plt.title('Number of Ramen Brands by Country')
plt.xlabel('Country')
plt.ylabel('Number of Brands')
plt.xticks(rotation=90)
```

Looking at this data, we can infer that Japan, United States, Taiwain, China, and South Korea are the top five contributors to the ramen brands out there. On the other hand Serbia, Israel, Peru, Portugal, and Nepal are the least contributors to the distribution of ramen brands. 

Next, we can take a look at which countries have ramen brands with higher ratings.

```python
final_ramen['Stars'] = pd.to_numeric(final_ramen['Stars'], errors='coerce')

# Calculate average ratings by country
average_ratings = final_ramen.groupby('Country')['Stars'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(14, 8))
average_ratings.plot(kind='bar', color='skyblue')
plt.grid(True)

# Customize the plot
plt.title('Average Ratings of Ramen Brands by Country')
plt.xlabel('Country')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```


From this visualization, we can determine that France, Malaysia, Cambodia, Indonesia, and Singapore are the top five countries with the highest rated ramen brands. We can observe that all five have ratings 4 and above.

On the other hand ramen brands from Nigeria, Canada, Serbia, Portugal, and Ireland are rated lower. We can observe that all five have ratings 2.5 and under. 

## 2. Flavor Variety

After coming to the conclusion that there is a difference in the number of ramen brands originating from different countries, we will now move on to analyzing wheter or not flavor variety contributes toward a ramen's rating.

#### Hypothesis
Null: There is a significance difference in ramen ratings depending on its key flavors.

Alternative: There is not a significance difference in ramen ratings depending on its key flavors.

First let's revisit the dataset we are using.

```python
final_ramen.head()
```


In order to analyze the ramen flavors, we've decided to perform an analysis using the Term Frequency and Inverse Document Frequency. Since the flavors are currently stored in a the column "Variety" we must first concatenate all of the flavors into one large text.

```python
flavor_text = final_ramen['Variety'].tolist()
flavor_text[:10]
```

```python
def preprocess_text(text):
    text = re.sub(r'\d+', '', text) 
    text = re.sub(r'[^a-zA-Z\s]', '', text)  
    text = text.lower() 
    return text

cleaned_flavor_text = [preprocess_text(words) for words in flavor_text]
cleaned_flavor_text[:10]
```

After we have created the text of all of the the flavors combined to only include english words (no foreign languages or numbers), we can now carry on and create a table of the TF-IDF frequencies.

```python
tfidf = TfidfVectorizer(sublinear_tf=True, analyzer='word', max_features=2000, stop_words='english')
matrix = tfidf.fit_transform(cleaned_flavor_text)  
flavor_tfidf = pd.DataFrame(matrix.toarray(), columns=tfidf.get_feature_names_out())
flavor_tfidf
```

The table above contains the TF-IDF of all of the words used in the "Variey" column, but it does not tell us much about how these word frequencies relate to the rating it gets. Therefore, we will create sub TF-IDF tables in relation to the ratings they get.

In order to do this, we will define a high rating as any entry that receives at least a 4.5 star rating.

```python

# create sub dataframe containing entries with at least a 4.5 star rating
ramen_high_rank = final_ramen[final_ramen['Stars'] >= 4.5]
ramen_high_rank.head()
```

```python
# cleaned list of strings containing the first 10 entries with a rating of at least 4.5
flavor_text_high = ramen_high_rank['Variety'].tolist()
cleaned_flavor_high = [preprocess_text(words) for words in flavor_text_high]
cleaned_flavor_high[:10]
```

```python
tfidf_high = TfidfVectorizer(sublinear_tf=True, analyzer='word', max_features=2000, stop_words='english')
matrix_high = tfidf_high.fit_transform(cleaned_flavor_high)  
flavor_tfidf_high = pd.DataFrame(matrix_high.toarray(), columns=tfidf_high.get_feature_names_out())
flavor_tfidf_high
```

```python
word_sums = flavor_tfidf_high.sum(axis=0)

# Sort the words by their total TF-IDF scores in descending order and get the top 10
top_30_words = word_sums.sort_values(ascending=False).head(30)

print("Top 30 most common words based on TF-IDF scores:")
print(top_30_words)
```

We notice that there are a few words that made it to the top 30 list that are not related to the flavor of the ramen so let's get rid of those in our final_top_words list which consists of the 20 top flavor words.

```python
top_words = top_30_words.index.tolist()
final_top_words = [
 'spicy',
 'soup',
 'curry',
 'chicken',
 'sauce',
 'tonkotsu',
 'beef',
 'yakisoba',
 'goreng',
 'penang',
 'rice',
 'miso',
 'udon',
 'seafood',
 'tom',
 'shoyu',
 'white',
 'shio',
 'pork',
 'yum']
```

Now that we have the TF-IDF dataframe, we will visualize these results using line graphs.

```python
pattern = '|'.join(final_top_words)

filtered_df = ramen_high_rank[ramen_high_rank['Variety'].str.contains(pattern, case=False, na=False)]

# Since there are too many variations of flavors, we will randomly choose 20 flavors to see the visualization 
random_sample_df = ramen_high_rank.sample(n=20, random_state=1)
random_sample_df
```

```python
star = random_sample_df['Stars'].to_numpy()
top_words = random_sample_df['Variety'].to_numpy()
plt.figure(figsize=(25, 6))
plt.plot(top_words, star, marker='o', linestyle='-')
plt.xticks(rotation=45)
f1 = plt.gcf()
```

Upon creating this line graph, we can see that these are the flavors that are typically rated with at least a 4.5 or higher due to their flavor name including one of the top 20 common key flavors.

After looking at the tfidf analysis, we will now perform the actual hypothesis test.

```python
contingency_table_variety = pd.crosstab(index=final_ramen['Stars'], columns=final_ramen['Variety'])

# Perform chi-square test for independence
chi2_statistic_1, p_value_1, dof, expected = chi2_contingency(contingency_table_variety)

# Interpret Results
if p_value_1 < alpha:
    print("Reject null hypothesis: There is a significance difference in ramen ratings depending on its key flavors.")
else:
    print("Fail to reject null hypothesis: There is no significance difference in ramen ratings depending on its key flavors.")
```

## 3. Packaging

After coming to the conclusion that there is a difference in ramen ratings depending on its key flavors, we will now move on to analyzing wheter or not flavor variety contributes toward a ramen's rating

#### Hypothesis
Null: There is significant difference in the ramen ratings based on packaging style.

Alternative: There is not a significant difference in the ramen ratings based on packaging style.

```python
contingency_table2 = pd.crosstab(index=final_ramen['Brand'], columns=final_ramen['Style'])


# Perform chi-square test for independence
chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table2)


# Interpret Results
alpha = 0.05
if p_value < alpha:
   print("Reject null hypothesis: There is significant difference in the ramen ratings based on packaging style.")
else:
   print("Fail to reject null hypothesis: There is no significant difference in the ramen ratings based on packaging style.")
```

We can then infer that a ramen's packaging style affects the overall rating of the ramen brand.

Lets make a visualization for the distribution of number of ramen per packaging style:

```python
ramen_counts_style = final_ramen.groupby('Style')['Brand'].nunique().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
colors2 = sns.color_palette("rocket", len(ramen_counts_style))
ramen_counts_style.plot(kind='bar', color=colors2)

plt.title('Number of Ramen Brands by Packaging')
plt.xlabel('Packaing Style')
plt.ylabel('Number of Brands')
plt.xticks(rotation=90)
```


Looking at this data, we can infer that the most popular packaging style is the 'Pack'.

Next, we can take a look at which packaging styles have ramen brands with higher ratings.

```python
final_ramen['Stars'] = pd.to_numeric(final_ramen['Stars'], errors='coerce')


# Calculate average ratings by country
average_ratings_style = final_ramen.groupby('Style')['Stars'].mean().sort_values(ascending=False)


# Plotting
plt.figure(figsize=(14, 8))
average_ratings_style.plot(kind='bar', color='red')
plt.grid(True)


# Customize the plot
plt.title('Average Ratings of Ramen Brands by Packaging Style')
plt.xlabel('Style')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability


# Show the plot
plt.show()	
```


From this visualization, we can conclude that the packaging style does influence a ramen brand's rating with 'Bar' usually resulting in a high average rating of 5 and a 'Pack' of ramen resulting in a average of 3.7
